```
class Serija:

  def __init__(self,naziv, vrednosti=[]):
    self.naziv=naziv
    self.vrednosti=vrednosti

  def prestej(self):
    return len(self.vrednosti)

  def opis(self):
    return sum(self.vrednosti)/len(self.vrednosti)

  def glava(self, x=4):
    l = []
    for i in self.vrednosti[0:x]:
      l.append(i)
      print(i)
    return l #vrne list stevilk glave 

  def rep(self,x=4):
    l =[]
    for i in self.vrednosti[-x:]:
      #print(i) 
      l.append(i)
      print(i)
    
    return l # vrne rep do 4 od zadi naprej


  def dodaj(self, vrednost, indeks=None):
    if indeks:
      self.vrednosti.insert(indeks,vrednost)
    else:
      self.vrednosti.append(vrednost)

  def zbrisi(self,indeks):
       del self.vrednosti[indeks]



class Razpredelnica(Serija):

   def __init__(self,stolpci=[]):
     self.stolpci=stolpci
  

   def prestej(self):
     a=[] # rezerviram novi list
     vrstica=1
     for i in self.stolpci: # da prestejem vse vrstice v stolpci
      if len(i.vrednosti) > vrstica:
        vrstica = len(i.vrednosti)

     print(vrstica)
     a.append(vrstica) #dodamo st vrstic
     a.append(len(self.stolpci)) #dodamo st stolpcev
     #print(a) 
     return a

   def opisi(self):
      #return sum(self.stolpec)/len(self.vrednosti)
      for i in self.stolpci:
       n =  print(i.naziv,":",sum(i.vrednosti)/len(i.vrednosti))
      return n
      
   def glava(self,x=3):
    for stolpec in self.stolpci:
         izpis_glave = stolpec.glava(x)
    return izpis_glave


   def rep(self,x=3):
    for stolpec in self.stolpci:
        izpis_repa = stolpec.rep(x)   
    return izpis_repa

   def dodaj_vrstico(self,vrstica, indeks=None):
       indeks=0
       for i in self.stolpci: # da gre po vseh stolpcah
          i.vrednosti.append(vrstica[indeks])
          indeks+=1
          print(i.naziv,i.vrednosti) # izpise
          if indeks == len(vrstica): # ce je indeks enak dolzini vrze ven
              break

   def zbrisi_vrstico(self,indeks): 
     for i in self.stolpci:
       del i.vrednosti[indeks]

   def dodaj_stolpec(self,stolpec, indeks=None):
     if indeks:
        self.stolpci.insert(indeks,stolpec)
     else:
        self.stolpci.append(stolpec)

   def zbrisi_stolpec(self,indeks):
       del self.stolpci[indeks]


s1 = Serija('visina', [1.5, 1.7, 1.4, 1.3, 1.9])
s2 = Serija('sirina', [5, 7, 4, 3, 9, 6])
s3 = Serija('starost', [15, 11, 10, 12])

#print(s1.naziv)

#VRACANJE VREDNOSTI
print(s1.prestej()) # 5
print(s2.prestej()) # 6
print(s3.prestej()) # 4

#za funkcijo opis pri za razred Serija
print(s1.opis())
print(s2.opis())
print(s3.opis())

#Funkija glava
s1.glava()



#Funkcija IZBRISI
s2.zbrisi(2)
print("TEST IZPIS GLAVE")
#Funkija glava
s1.glava()
print("---KONEC GLAVE")
#Funkcija dodaj
print("TESTIRAM")
s3.dodaj(13, indeks=1)

print(s1.prestej()) # 5
print(s2.prestej()) # 5
print(s3.prestej()) # 5
#NOVI CLASS
print("novi class")
raz = Razpredelnica([s1, s2, s3])

print(raz.prestej()) # [5, 3]
print("HELLO")

print("VRNE GLAVO")
raz.dodaj_vrstico([1.6, 8, 14])
raz.dodaj_stolpec(Serija('teza', [34, 21, 56, 43, 51, 28]))
print(raz.prestej()) # + 1
print("BRISANJE STOLPCEV")
#ZA GLAVO IN REP OD RAZPREDELNICE
print("<--- ZACETEK GLAVE")
raz.glava()
print("<--- KONEC GLAVE")

#FUNKCIJA ZA REP
print("------BRANJE REPA")
raz.rep()
print("------KONEC BRANJA REPA")
#PRIKAZ POVPRECJA STOLPCA
#print(raz.opisi()) #
raz.zbrisi_vrstico(1)
raz.zbrisi_stolpec(0)

print(raz.prestej()) # + 1

#Funkija rep
s1.rep()


```
